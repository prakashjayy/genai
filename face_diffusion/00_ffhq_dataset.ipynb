> Lets understand FFHQ dataset

```python
import numpy as np 
import fastcore.all as fc
import PIL
from PIL import Image


import matplotlib.pyplot as plt
plt.style.use("bmh")
%matplotlib inline 
```

```python
x = fc.L(fc.Path("/root/prakash/data/ffhq/kaggle-images/").glob("*.png"))
x
```

```python
def image_grid(imgs, rows, cols):
    w,h = imgs[0].size
    grid = Image.new('RGB', size=(cols*w, rows*h))
    for i, img in enumerate(imgs): grid.paste(img, box=(i%cols*w, i//cols*h))
    return grid

def thumbnail(img, size=256):
    if not isinstance(img, PIL.Image.Image): img = Image.fromarray(img)
    w, h = img.size
    ar = h/w 
    return img.resize((size, int(size*ar)))
```

```python
imgs = [Image.open(x[np.random.randint(len(x))]) for i in range(16)]
thumbnail(image_grid(imgs, 4, 4), 256)
```
