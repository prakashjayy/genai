what are vector fields?

there are several definitions to it. but one way is of showing 3D objects in 2D dimensions

$$
\mathbf{f}(x, y) = \begin{bmatrix} y^3-9y \\ x^3-9x \end{bmatrix}
$$

lets solve the above equation at (1, 1) and we get (-8, -8). this can be visualized below 




```python
# first have a grid -8, 8 range 

plt.figure(figsize=(4, 4))
plt.quiver(1, 1, -9, -9, angles='xy', scale_units='xy', scale=1, width=0.005, headwidth=3, color='red')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()



```

```python
# Create a grid of points
x = np.linspace(-6, 6, 20)
y = np.linspace(-6, 6, 20)
X, Y = np.meshgrid(x, y)

# Calculate vector components
# f(x, y) = [[y続-9y], [x続-9x]]
U = Y**3 - 9*Y  # x-component: y続-9y
V = X**3 - 9*X  # y-component: x続-9x
```

> if u want to scale the length of the vector between 0 and 1 

> color represent their magnitude in size.

```python
# Calculate vector magnitudes for color mapping
magnitude = np.sqrt(U**2 + V**2)


#norm_factor = 0.8 * np.mean(magnitude)
#U_norm = U / (magnitude + 1e-10) * norm_factor
#V_norm = V / (magnitude + 1e-10) * norm_factor
U_norm = U/U.max()
V_norm = V/V.max()

# Create the plot
plt.figure(figsize=(5, 5))

# Plot the vector field with color based on magnitude
plt.quiver(X, Y, U_norm, V_norm, magnitude, cmap='viridis', 
           pivot='mid', angles='xy', scale_units='xy', scale=1)
plt.show()
```

> color still represent their magnitude but all vectors are of same length

```python
# Calculate vector magnitudes for color mapping
magnitude = np.sqrt(U**2 + V**2)


#norm_factor = 0.8 * np.mean(magnitude)
#U_norm = U / (magnitude + 1e-10) * norm_factor
#V_norm = V / (magnitude + 1e-10) * norm_factor
U_norm = U/magnitude
V_norm = V/magnitude

# Create the plot
plt.figure(figsize=(5, 5))

# Plot the vector field with color based on magnitude
plt.quiver(X, Y, U_norm, V_norm, magnitude, cmap='viridis', 
           pivot='mid', angles='xy', scale_units='xy', scale=1)
plt.show()
```

## contour lines 

$$
f(x, y) = xy
$$



```python
# Create a grid of points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X*Y
```

```python
fig, ax = plt.subplots(figsize=(5, 4))

# Create contour plot
contour = ax.contour(X, Y, Z, levels=25, cmap='viridis')

# Add contour labels
plt.clabel(contour, inline=True, fontsize=10, fmt='%1.1f')

# Add a colorbar
plt.colorbar(contour, label='f(x,y) = xy')
```
